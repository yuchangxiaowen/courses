```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python
def plot(Hs):
    means = [H.mean() for H in Hs]
    stds = [H.std() for H in Hs]

    plt.plot(means, "or-")
    plt.show()
    plt.plot(stds, "ob-")
    plt.show()

    for i, H in enumerate(Hs):
        if i // 3 * 3 == i:
            plt.figure()
        plt.subplot(1, 3, i % 3 + 1)
        plt.hist(H.ravel(), 30, range=(-1, 1))
    plt.show()
```

```python
inputs = np.random.randn(1000, 500)
hidden_layer_sizes = [500] * 9
```

```python
Hs = []
for i in range(len(hidden_layer_sizes)):
    X = inputs if i == 0 else Hs[i - 1]
    fan_in = X.shape[1]
    fan_out = hidden_layer_sizes[i]
    W = np.random.randn(fan_in, fan_out)
    Hs.append(np.tanh(X.dot(W)))

plot(Hs)
```

```python
Hs = []
for i in range(len(hidden_layer_sizes)):
    X = inputs if i == 0 else Hs[i - 1]
    fan_in = X.shape[1]
    fan_out = hidden_layer_sizes[i]
    W = np.random.randn(fan_in, fan_out) / math.sqrt(fan_out)
    Hs.append(np.tanh(X.dot(W)))

plot(Hs)
```
