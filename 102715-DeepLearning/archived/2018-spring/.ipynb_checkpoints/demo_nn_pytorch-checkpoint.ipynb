```python
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self, input_sz, hidden_sz, output_sz):
        # Step 1: params store
        super(Net, self).__init__()
        self._input_sz = input_sz
        self._hidden_sz = hidden_sz
        self._output_sz = output_sz
        # Step 2: layer definition
        self._fc1 = nn.Linear(input_sz, hidden_sz[0])
        self._fc2 = nn.Linear(hidden_sz[0], hidden_sz[1])
        self._fc3 = nn.Linear(hidden_sz[1], output_sz)

    def forward(self, x):
        x = F.tanh(self._fc1(x))
        x = F.tanh(self._fc2(x))
        x = F.softmax(F.tanh(self._fc3(x)), dim=1)
        return x
```

```python
import numpy as np
import torch
import matplotlib.pyplot as plt


def load_data(train, batch_size, is_show=False):
    n = 600 if train else 100
    name = "train" if train else "test"
    theta = torch.rand(n) * 2 * np.pi
    tmp = torch.rand(n) * 2
    r = torch.sqrt(tmp + (tmp > 1).type(torch.FloatTensor))
    x1, x2 = r * np.sin(theta), r * np.cos(theta)
    inputs = torch.split(
        torch.cat((x1.unsqueeze(1), x2.unsqueeze(1)), 1),
        batch_size, 0)
    labels = torch.split(
        (r <= 1).type(torch.LongTensor),
        batch_size, 0)
    if is_show:
        show_data(inputs, labels, name)
    return list(zip(inputs, labels))


def show_data(x, y, name):
    x, y = torch.cat(x), torch.cat(y)
    plt.plot(x[:, 0][y.eq(0)].numpy(),
        x[:, 1][y.eq(0)].numpy(), "*")
    plt.plot(x[:, 0][y.eq(1)].numpy(),
        x[:, 1][y.eq(1)].numpy(), "*")
    plt.show()
```

```python
import torch.optim as optim
from torch.autograd import Variable


use_cuda = False
input_sz = 2
hidden_sz = (4, 2)
output_sz = 2

# build nn
net = Net(
    input_sz=input_sz,
    hidden_sz=hidden_sz,
    output_sz=output_sz
)
if use_cuda:
    net.to(torch.device("cuda"))

# input config
batch_sz = 10
trainloader = load_data(train=True, batch_size=batch_sz, is_show=True)
testloader = load_data(train=False, batch_size=batch_sz, is_show=True)

# loss func and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=.03)
```

```python
# training
num_epoch = 50
losses = []
accs = []
for epoch in range(num_epoch):
    running_loss = .0
    for i, data in enumerate(trainloader, 0):
        inputs, labels = data
        inputs, labels = ((
            Variable(inputs.to(torch.device("cuda"))),
            Variable(labels.to(torch.device("cuda"))))
            if use_cuda else (
                Variable(inputs),
                Variable(labels)))
        # grad buffer clear
        optimizer.zero_grad()
        outputs = net(inputs)
        # shape of outputs: (n_samples, n_classes)
        # shape of labels: (n_samples)
        loss = criterion(outputs, labels)
        # bp with sgd
        loss.backward()
        # update value
        optimizer.step()
        running_loss += loss.data
    losses.append(running_loss / (i + 1))
#     print("Epoch: {}, training loss: {}".format(
#         epoch + 1, losses[-1]))
    # test set performance
    idx = epoch % (100 // batch_sz)
    inputs, labels = testloader[idx]
    outputs = net(Variable(inputs))
    _, pred = torch.max(outputs.data, 1)
    acc = (pred == labels).sum() / labels.size(0)
    accs.append(acc)
```

```python
plt.plot(range(1, len(accs) + 1), accs)
plt.title("accs")
plt.show()
```
